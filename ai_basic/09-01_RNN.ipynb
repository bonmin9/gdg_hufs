# 09. 텍스트를 위한 인공 신경망

# 09-01.순차 데이터와 순환 신경망

## 순차 데이터(sequential data)
- 텍스트 혹은 시계열 데이터 등 순서에 의미가 있는 데이터
- 완전 연결 신경망과 합성곱 신경망은 데이터의 흐름이 앞으로만 전달되는 피드포워드 신경망(feedforward neural network)이기 때문에 이전 입력 데이터를 기억해야하는 순차 데이터 처리에는 적합하지 않음

## 순환 신경망(recurrent neural network, RNN)
- 뉴런의 출력이 다시 자기 자신에게 전달되어 다음 입력에 영향을 미치는 신경망 알고리즘

### 용어 정리
- 셀(cell): 순환 신경망에서 층을 가리키는 말로 뉴런을 일일이 표시하지 않고 하나의 셀로 층을 표현
- 은닉 상태(hidden state): 셀의 출력
- 시퀀스(sequence): 순환 신경망의 샘플로 대부분 순차적인 데이터
- 타임스텝(timestep): 샘플을 처리하는 한 단계로 타임스텝의 길이는 시퀀스의 길이와 동일
  - 텍스트 데이터라면 타임스텝의 길이는 한 문장(시퀀스)의 단어 개수가 됨

### 순환 신경망의 기본 구조
1. 첫 번째 타임스텝에서는 단순하게 입력에 가중치 $w_x$를 곱하고 활성화함수를 적용하여 은닉 상태를 출력한다.
2. 은닉 상태를 자기 자신을 포함한 같은 층의 모든 뉴런에 전달한다.
3. 두 번째 타임스텝부터는 기존값에 이전 타임스텝의 은닉 상태 $h$와 가중치 $w_h$를 곱한 값을 더한 후 활성화 함수를 적용해 은닉 상태를 출력한다. $w_h$는 모든 타임스텝에서 동일하다.
4. 2로 돌아가 반복한다.

- 이러한 과정을 거치면 이전 은닉 상태가 사라지지 않고 다음 은닉 상태에 영향을 주게되며 최종 은닉 상태는 모든 이전 은닉 상태에 영향을 받게 된다.

#### 하이퍼볼릭 탄젠트(hyperbolic tangent)
- 순환층에서 일반적으로 사용하는 활성화 함수로 tanh으로 표기한다.
- -1~1 사이의 값을 범위로 가진다.

### 셀의 파라미터 개수
- 입력층과 순환층을 이어주는 가중치의 개수는 특성 개수 x 순환층의 뉴런 개수
  - 특성이 4개이고 뉴런이 3개라면 12개
- 타임스텝마다 은닉 상태는 자기 자신을 포함한 모든 뉴런에 전달되어야하고 각각 가중치를 가지고 있어야 하므로 은닉 상태에 곱해지는 가중치의 개수는 $뉴런의 개수^2$
  - 뉴런의 개수가 3개라면 9개
- 각 뉴런 당 하나의 절편
  - 뉴런의 개수가 3개라면 3개
> 12 + 9 + 3 = 24

### 순환 신경망의 입출력(텍스트 데이터)
- 순환층의 시퀀스는 일반적으로 2개의 차원을 가진다.   
한 문장이 하나의 시퀀스를 구성한다면 각 단어가 첫 번째 차원에 배치되고 단어를 표현하는 일련의 숫자 배열이 2번째 차원을 구성한다.
- 시퀀스의 길이, 즉 첫 번째 차원의 길이만큼 타임스텝이 진행된다.
- 순환층은 마지막 은닉 상태만 반환한다.   
모든 타임스텝에 대한 결과가 하나의 은닉 상태로 출력되기 때문에 결과값은 1차원 배열이 된다.
- 은닉 상태는 뉴런의 개수만큼 주어지므로 1차원 배열의 크기는 뉴런의 개수와 동일하다.
- 순환 신경망도 마지막에 밀집층을 두어 클래스를 분류하지만 이미 최종 출력이 1차원 배열이기 때문에 Flatten 클래스를 사용할 필요는 없다.

#### 순환 신경망이 2개 이상 이어진다면?
- 순환층은 2차원 배열을 입력으로 받기 때문에 마지막 순환층을 제외한 순환층 역시 2차원 배열을 반환해야 한다.
